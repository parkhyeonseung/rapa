```python
import urllib.request
import ssl
context = ssl._create_unverified_context()
#환율정보 페이지 불러오기
print("[환율정보 불러오기]")
URL = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%ED%99%98%EC%9C%A8"
page = urllib.request.urlopen(URL, context=context)
text = page.read().decode("utf8")  # 해당 페이지의 소스코드
# print(text)
#환율정보 가져오기: 기준시간
searchText = 'class="grp_info"> <em>'
where      = text.find(searchText)  #소스코드에서 해당 문자열의 시작위치
targetInfo = text[where+len(searchText):where+len(searchText)+16]
print(f'-환율기준시간: {targetInfo}')
#환율정보 가져오기: 달러
searchText = '<span>미국 <em>USD</em></span></a></th> <td><span>'
where      = text.find(searchText)
targetInfo = text[where+len(searchText):where+len(searchText)+8]
print(f'-달러: {targetInfo}')
#환율정보 가져오기: 위안
searchText = '<span>중국 <em>CNY</em></span></a></th> <td><span>'
where      = text.find(searchText)
targetInfo = text[where+len(searchText):where+len(searchText)+6]
print(f'-위안: {targetInfo}')
#환율정보 가져오기:유로화
searchText = '<span>유럽연합 <em>EUR</em></span></a></th> <td><span>'
where      = text.find(searchText)
targetInfo = text[where+len(searchText):where+len(searchText)+8]
print(f'-유로: {targetInfo}')

```

```python
a = input('괄호 :')
x = list(a)  ## list
i=0          ## 초기값
while 1:
    if x[0]=='(' and x[-1]==')':    ## 맨앞과 맨뒤가 다른지
        if x[i]!=x[i+1]:             ## 앞과 뒤가 다른지
            del x[i:i+2]             ## 그러면 그 두 원소 삭제
            i=0                       ## 초기값 재설정
        
            if len(x)==0:             ## list x 가 비었는지
                print('True')
                break   
            elif len(x)==1:           ##        원소가 하나인지
                print('False')
                break
        else :                         ## 먄앞과 맨뒤가 다르지 않으면
            i+=1                        ## 초기값 +1
    else:
        print('False')
        break
```

```python
menu = '''[커피 자동주문 머신 메뉴]
---------------------------------------------
-아메리카노 2500원
-카페라떼   3000원
-카푸치노   3000원
입니다. 원하시는 커피종류와 잔수를 입력하세요.
'''
cof1 = int(input(menu+'아메리카노 몇잔?'))
cof2 = int(input('카페라떼 몇잔?'))
cof3 = int(input('카푸치노 몇잔?'))

all = 2500*cof1+3000*cof2+3000*cof3

print(f'지불할 총 금액은 {all}원 입니다.')
put_c = int(input(' 돈을 넣어 주세요~'))
while 1 :
    if put_c<all:
        put_mcc = all-put_c
        print(f'입력한 금액이 부족합니다. {put_mcc}원을 더 넣어주세요.')
        put_mc = int(input())
        print(f'{put_c}원 받았습니다.')
        put_c +=put_mc

    else:
        rest_c = put_c - all
        print(f'거스름돈은 {rest_c}원 입니다.')
        break
```

### Random

```python
import random
print(random.randint(10,20))           ## 10 이상 20 이하 정수
print(random.randrange(10,20,2))       ## 10 이상 20 미만 2스텝 정수
print(random.random())                 ## 0 ~1 실수
print(random.uniform(10,20))           ## 10 이상 20 이하 실수
```

```python
a_list = list(range(1,11))

random.shuffle(a_list)
print(a_list)
random.choice(a_list)
```

```python
box = list(range(1,7))
a_choice = random.choice(box)
b_choice = random.choice(box)
if a_choice >b_choice:
    print('A가 이겼습니다.')
elif a_choice<b_choice:
    print('B가 이겼습니다.')
else:
    print('비겼습니다.')
```

```python
A = [random.choice(box) for _ in range(2)]   ## 2개 뽑기 (중복 O)
print(A)
```

```python
a = random.choices(box,k=2)       #### 2개 뽑기 (중복 O)
print(a)
b = random.sample(box,k=2)        #### 2개 뽑기 (중복 X)
print(b)
```

```python
n = int(input('합계를 구할 숫자 입력:'))
sum=0
for i in range(1,n+1):
    if i%2==1:
        sum+=i
print(f'1~{n}까지 짝수를 제외한 합은: {sum}')
```

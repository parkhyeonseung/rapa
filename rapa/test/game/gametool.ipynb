```python
from tkinter import *
from tkinter.simpledialog import *
from tkinter import messagebox
from PIL import ImageTk, Image
import random
import urllib.request
import ssl
from collections import Counter
import time

def lotto_update():
    num_list = list(range(1,46))
    num_lotto = sorted(random.sample(num_list ,6))
    num_bonus = random.choice(num_lotto)
    bonus_i = num_lotto.index(num_bonus)
    del num_lotto[bonus_i]
    label1.configure(text=f'!!이번주 1등 로또 번호는!!\n{num_lotto}\nBonus Number : {num_bonus}\n입니다.',font=('bold',50),fg='blue')
    
def lotto_count():
    global all_c,text
    window2 = Tk()
    window2.title('lotto')
    window2.resizable(width=TRUE,height=TRUE)
    window2.geometry('700x700-800+200')
    
    context = ssl._create_unverified_context()
    label2 = Label(window2,text = 'lotto 1등 번호 빈도수',font=('bold',30))
    label2.pack(side=TOP)
    
    num = askinteger('lotto','총 몇 회차의 번호의 횟수을 알고싶나요?',minvalue = 1,maxvalue=100)
    
    time.sleep(0.5)
    
    all_num = []
    all_c = []
    for i in range(num):
        num = 979-i
        num = str(num)
        URL = "https://www.dhlottery.co.kr/common.do?method=getLottoNumber&drwNo="+num
        page = urllib.request.urlopen(URL, context=context)
        text = page.read().decode("utf8")  # 해당 페이지의 소스코드
        ta_na = '"drwNo":'
        ta_idx = text.find(ta_na)
        ta = text[ta_idx+len(ta_na):ta_idx+len(ta_na)+3]
        all_c.append(ta)
        for k in range(1,7):
            k=str(k)
            target_name = 'drwtNo'+k+'":'
            
            target_idx = text.find(target_name)
            
            target = text[target_idx+len(target_name):target_idx+len(target_name)+2]
            
            all_num.append(target)
    c_len = len(all_c)        
    labell3 = Label(window2,text = f'회차 : {all_c[0]}회 ~ {all_c[c_len-1]}회',font=('bold',15))
    labell3.pack(side=TOP)
    
    for i,v in enumerate(all_num):
        if ',' in v:
            v = v[0]
            all_num[i] = v
        elif '}' in v:
            v = v[0]
            all_num[i] = v

    counter_lot = Counter(all_num)
    counter_lot = sorted(counter_lot.items(), key=lambda x: x[1], reverse=True)
    co=0
    if int(num) <979:
        while counter_lot[5+co][1] == counter_lot[6+co][1]:
            co+=1
    labell2 = [None]*co
    labell = [None]*6

    for i in range(6):
        labell[i]=Label(window2,text = f'숫자 : {counter_lot[i][0]}\t| 횟수 : {counter_lot[i][1]}',font=('bold',15),fg='blue')
        labell[i].pack(side=TOP)
    if co!=0:
        labell3 = Label(window2,text='그 외의 동률 숫자 :',font=('bold',20))
        labell3.pack(side=LEFT)
        for ro in range(co):
            labell2[ro] = Label(window2,text = f' 숫자 : {counter_lot[6+ro][0]} -- 횟수 : {counter_lot[6+ro][1]}  // ',font=('bold',13))
            labell2[ro].pack(side=LEFT)
    else :
        pass

    window2.wm_attributes("-topmost", 1)
    window2.mainloop()

    
    
def lotto():
    global num_lotto,label1,num_bonus
    window1=Tk()
    window1.title('lotto')
    window1.resizable(width=TRUE,height=TRUE)
    window1.geometry('700x580')
    num_list = list(range(1,46))
    num_lotto = sorted(random.sample(num_list ,7))
    num_bonus = random.choice(num_lotto)
    bonus_i = num_lotto.index(num_bonus)
    del num_lotto[bonus_i]
    label2 = Label(window1,text=f'$$$$$$$$$$$$$$$$',font=('bold',60),fg='red')
    label1 = Label(window1,text=f'!!이번주 1등 로또 번호는!!\n{num_lotto}\nBonus Number : {num_bonus}\n입니다.',font=('bold',50),fg='blue')
    label1.place(y=130)
    label2.place(y=30)
    button1 = Button(window1,text='맘에 안들어?',bg='white',command = lotto_update )
    button1.place(x=250,y=520)
    window1.mainloop()
    
def twenty():
    messagebox.showinfo('스무고개','비밀의 숫자가 지정됨!')
    bar='-'
    num = random.randrange(1,101)
    put_num = 0
    count = 0
    while 1:
        try:
            put_num = askinteger('스무고개','숫자를 입력하세요',minvalue=1,maxvalue=100) 
        except :
            press = messagebox.askyesno('스무고개','그만 둘건가요?')
            if press1 ==True:
                break
            elif press1 == False:
                pass
        count+=1
        try:
            if put_num<num:
                messagebox.showinfo('스무고개','비밀의 수가 더 커!')
            elif put_num>num:
                messagebox.showinfo('스무고개','비밀의 수가 더 작아!')
            else :
                break
        except:
            press = messagebox.askyesno('스무고개','그만 둘건가요?')
            if press ==True:
                break
            elif press == False:
                pass
    if press ==True:
        messagebox.showinfo('스무고개','비밀의 수 : 안알랴줌^^')
        messagebox.showinfo('스무고개',f'시도 횟수 : {count}')
    else:
        messagebox.showinfo('스무고개',f'정답 !!!\n비밀의 수 : {num}')
        messagebox.showinfo('스무고개',f'시도 횟수 : {count}')

def rock():
    global rsp_u
    while 1:
        rsp_ai_dict = {0:'가위', 1:'바위', 2:'보'}       ## ai 딕
        rsp_u_dict = {'가위':0, '바위':1, '보':2}        ## 사용자 딕

        rsp_ai_num = random.choice(list(range(3)))      ## ai 무기 번호 고르기  int
        rsp_ai = rsp_ai_dict[rsp_ai_num]                ## ai 무기 번호로 무기 선택 str
        
        try:
            rsp_u = askstring('rock','가위~ 바위~ 보!\n당신의 무기를 입력하세요! :')   ## 사용자 무기 입력  str

            if rsp_u in ['가위','바위','보']:

                if rsp_u_dict[rsp_u] == rsp_ai_num:
                    result = 'draw'

                elif rsp_u_dict[rsp_u] == 0 and rsp_ai_num == 2:
                        result = 'win'
                elif rsp_u_dict[rsp_u] == 1 and rsp_ai_num == 0:
                        result = 'win'
                elif rsp_u_dict[rsp_u] == 2 and rsp_ai_num == 1:
                        result = 'win'

                else:
                    result = 'loose'
    
            else :
                messagebox.showinfo('rock','가위, 바위, 보 중에 선택하세요!')
                quit1 = messagebox.askyesno('quit','그만?')
                if quit1 == True:
                    break
                else :
                    continue
         
        except:
            quit = messagebox.askyesno('retry?','그만? ')
            if quit == True:
                break
            elif quit == False :
                continue
            else :
                break
                
        messagebox.showinfo('rock',f'Ai 무기: {rsp_ai} VS 너의 무기: {rsp_u}\n| You {result}!!! |')
        
        while 1:
            repeat = messagebox.askyesno('retry?','한번더? ')
            if repeat == True:
                break
            else :
                break
                
        if repeat == True :
            continue
        elif repeat == False:
            messagebox.showinfo('rock','게임 끝')
            break

def count():
    instr = askstring('word','글자 빈도수를 알고 싶은 글을 적어봐!: ')
    counter = Counter(instr)  # 딕셔너리타입
    
    del counter[' ']
    n = len(counter)
    count_key = list(counter.keys())
    count_val = list(counter.values())
    window4 = Tk()
    window4.title('count')
    window4.resizable(width=TRUE,height=TRUE)
    window4.geometry('700x580')
    label1 = [None]*n
    label2 = Label(window4,text = f'입력된 문장 : {instr}',anchor = NW,font=('bold',20))
    label2.pack()
    label = Label(window4,text = '글자 빈도수:\n', font=('bold',40),fg='blue' )
    label.pack(side=TOP)
    for i in range(n):
        label1[i] = Label(window4,text = f'{count_key[i]} : {count_val[i]} 번',font=('bold',15),fg='blue')
        label1[i].pack(side = TOP)
    window4.mainloop()
            
img_path = 'games/'

window = Tk()

window.title('미니 게임')
window.geometry('1000x600')
window.resizable(width=FALSE,height=FALSE)


photo1 = ImageTk.PhotoImage(Image.open(img_path+'lotto.png'))
button1_1 = Button(image = photo1,command=lotto)

photo2 = ImageTk.PhotoImage(Image.open(img_path+'twenty.jpeg'))
button2_1 = Button(image = photo2, command=twenty)

photo3 = ImageTk.PhotoImage(Image.open(img_path+'rock.jpeg'))
button3_1 = Button(image = photo3, command = rock)

photo4 = ImageTk.PhotoImage(Image.open(img_path+'count.jpeg'))
button4_1 = Button(image = photo4, command = count)

button1_2 = Button(text = '1등 번호 data',command = lotto_count)

button1_1.place(x=80,y=100)
button2_1.place(x=600,y=50)
button3_1.place(x=150,y=320)
button4_1.place(x=560,y=340)

button1_2.place(x=80,y=20)

window.mainloop()
```
